```python
import matplotlib.pyplot as plt # Thư viện
from scipy import signal 
import numpy as np
```

```python
# Tín hiệu gốc
fm = 3
t = np.arange(0, 1, 0.001)
x = np.sin(2*np.pi*fm*t)
```

```python
# Lấy mẫu với tần số thấp
fs_low = 4
t_sampledlow = np.arange(0, 1, 1/fs_low)
x_sampledlow = np.sin(2*np.pi*fm*t_sampledlow)
```

```python
# Lấy mẫu với tần số cao
fs_high = 10
t_sampledhigh = np.arange(0, 1, 1/fs_high)
x_sampledhigh = np.sin(2*np.pi*fm*t_sampledhigh)
```

```python
# Vẽ
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(t, x)
plt.title('Original Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.stem(t_sampledlow, x_sampledlow, 'r')
plt.title('Sampling Signal with Low Frequency')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.stem(t_sampledhigh, x_sampledhigh, 'g')
plt.title('Sampling Signal with High Frequency')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()
```

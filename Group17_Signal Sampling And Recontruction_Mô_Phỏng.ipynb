```python
import matplotlib.pyplot as plt # Thư viện
from scipy import signal 
import numpy as np
```

```python
# Tín hiệu có dạng: 
x = 5*sin(2*np.pi*10*t)
fmax = 20pi/2pi = 10
Theo định lý lấy mẫu nyquist: fs >= 2fmax = 20Hz
```

```python
t = np.linspace(0, 1, 1000, endpoint=False)
x = 5*np.sin(2*np.pi*10*t)

'''
Tín hiệu gốc
'''
plt.plot(t, x, 'r')
plt.title('Original Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

```python
'''
Lấy mẫu từng fs
'''
fs1 = 10  # Trường hợp fs < 2fmax --> aliasing
fs2 = 20  # fs = 2fmax
fs3 = 500 # fs > 2fmax

ts1 = 1/fs1
ts2 = 1/fs2
ts3 = 1/fs3

t_sampled1 = np.arange(0, 1, ts1)
t_sampled2 = np.arange(0, 1, ts2)
t_sampled3 = np.arange(0, 1, ts3)

x_sampled1 = np.sin(2*np.pi*10*t_sampled1)
x_sampled2 = np.sin(2*np.pi*10*t_sampled2)
x_sampled3 = np.sin(2*np.pi*10*t_sampled3)

plt.stem(t_sampled1, x_sampled1, 'b')
plt.title('Sampled Signal fs1')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()

plt.stem(t_sampled2, x_sampled2, 'b')
plt.title('Sampled Signal fs2')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()

plt.stem(t_sampled3, x_sampled3, 'b')
plt.title('Sampled Signal fs3')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

```python
'''
Khôi phục tín hiệu của từng fs
'''
x_reconstructed_1 = np.interp(t, t_sampled1, x_sampled1) 

plt.plot(t, x_reconstructed_1, 'g')
plt.title('Reconstructed Signal fs1')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()

x_reconstructed_2 = np.interp(t, t_sampled2, x_sampled2) 

plt.plot(t, x_reconstructed_2, 'g')
plt.title('Reconstructed Signal fs2')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()

x_reconstructed_3 = np.interp(t, t_sampled3, x_sampled3) 

plt.plot(t, x_reconstructed_3, 'g')
plt.title('Reconstructed Signal fs3')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

```python
Nhận xét: 
    - Tần số lấy mẫu càng lớn thì tín hiệu khôi phục càng giống với tín hiệu gốc
```

```python
import numpy as np # Thư viện
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.signal import resample
```

```python
# 3.1. Thiết Lập Mô Phỏng
# Tạo một tín hiệu phức tạp (ví dụ: hai hàm sin với tần số khác nhau)
fs = 1000  # Tần số lấy mẫu
t = np.arange(0, 1, 1/fs)  # Thời gian từ 0 đến 1 giây
signal_original = np.sin(2 * np.pi * 5 * t) + 0.5 * np.sin(2 * np.pi * 20 * t)  # Tín hiệu phức tạp

# 3.2. Sử Dụng Mô Hình Tín Hiệu
signal_model = signal_original + 0.2 * np.random.randn(len(t))  # Add noise to the original signal

# 3.3. Mẫu Tín Hiệu với Các Tỉ Lệ Khác Nhau
# Tạo các tín hiệu mẫu với tỉ lệ khác nhau
downsampling_factor_low = 5
downsampling_factor_medium = 10
downsampling_factor_high = 20

signal_low = resample(signal_model, len(t)//downsampling_factor_low)
signal_medium = resample(signal_model, len(t)//downsampling_factor_medium)
signal_high = resample(signal_model, len(t)//downsampling_factor_high)

# 3.4. Áp Dụng Các Phương Pháp Khôi Phục
# Sử dụng các phương pháp khác nhau như Zero-Order Hold, Linear Interpolation, Sinc Interpolation
def zero_order_hold_interpolation(original, factor):
    # Áp dụng Zero-Order Hold
    resampled = np.repeat(original, factor)
    return resampled[:len(original)]

def linear_interpolation(original, factor):
    # Áp dụng Linear Interpolation
    f = interp1d(np.arange(0, len(original)), original, kind='linear')
    resampled = f(np.linspace(0, len(original)-1, len(original)*factor))
    return resampled

def sinc_interpolation(original, factor):
    # Áp dụng Sinc Interpolation
    resampled = resample(original, len(original)*factor)
    return resampled

def regression_interpolation(original, factor):
    # Áp dụng phương pháp hồi quy
    x_original = np.arange(0, len(original))
    x_resampled = np.linspace(0, len(original)-1, len(original)*factor)
    coeffs = np.polyfit(x_original, original, deg=2)
    resampled = np.polyval(coeffs, x_resampled)
    return resampled

# Chọn phương pháp khôi phục
reconstruction_methods = [zero_order_hold_interpolation, linear_interpolation, sinc_interpolation, regression_interpolation]

# 3.5. Đánh Giá Chất Lượng Tín Hiệu Tái Tạo
# Đánh giá chất lượng bằng cách so sánh tín hiệu tái tạo với tín hiệu gốc, tính độ nhiễu, và độ méo.

# Định nghĩa hàm đánh giá chất lượng tín hiệu
def evaluate_signal_quality(original, reconstructed, method_name, t_reconstructed):
    # Hiển thị tín hiệu gốc và tín hiệu tái tạo
    plt.figure(figsize=(10, 6))
    plt.plot(t, original, label='Original Signal', alpha=0.7)
    plt.plot(t_reconstructed, reconstructed, label=f'Reconstructed Signal ({method_name})', linestyle='dashed')
    plt.title(f'Comparison between Original and Reconstructed Signals ({method_name})')
    plt.legend()
    plt.show()

    # Tính độ nhiễu
    noise = original[:len(t_reconstructed)] - reconstructed
    noise_power = np.sum(noise**2)
    
    # Tính độ méo
    distortion = np.sum((original[:len(t_reconstructed)] - reconstructed)**2) / np.sum(original[:len(t_reconstructed)]**2)
    distortion_db = 10 * np.log10(distortion)

    print(f'Đánh Giá Chất Lượng Tín Hiệu Tái Tạo ({method_name}):')
    print(f' - Độ Nhiễu: {noise_power:.4f}')
    print(f' - Độ Méo: {distortion_db:.4f} dB\n')

# Áp dụng hàm đánh giá chất lượng cho từng phương pháp khôi phục
for method in reconstruction_methods:
    reconstructed_low = method(signal_low, downsampling_factor_low)
    reconstructed_medium = method(signal_medium, downsampling_factor_medium)
    reconstructed_high = method(signal_high, downsampling_factor_high)

    t_reconstructed_low = np.linspace(0, 1, len(reconstructed_low))
    t_reconstructed_medium = np.linspace(0, 1, len(reconstructed_medium))
    t_reconstructed_high = np.linspace(0, 1, len(reconstructed_high))

    evaluate_signal_quality(signal_original, reconstructed_low, f'Low Downsampling - {method.__name__}', t_reconstructed_low)
    evaluate_signal_quality(signal_original, reconstructed_medium, f'Medium Downsampling - {method.__name__}', t_reconstructed_medium)
    evaluate_signal_quality(signal_original, reconstructed_high, f'High Downsampling - {method.__name__}', t_reconstructed_high)
```
